# 03 - Modelagem e Avaliação

```python
import sys
sys.path.append('../src')

import pandas as pd
from train import get_models, train_model, save_model
from evaluate import evaluate_model, plot_confusion_matrix, compare_models
```

```python
X_train = pd.read_csv('../data/processed/X_train.csv')
X_test = pd.read_csv('../data/processed/X_test.csv')
y_train = pd.read_csv('../data/processed/y_train.csv').squeeze()
y_test = pd.read_csv('../data/processed/y_test.csv').squeeze()
```

```python
models = get_models()
results = {}
trained_models = {}

for name, model in models.items():
    print(f'--- Treinando {name} ---')
    model = train_model(model, X_train, y_train)
    trained_models[name] = model
    metrics = evaluate_model(model, X_test, y_test)
    results[name] = metrics
    print()
```

```python
compare_models(results)
```

```python
# Matriz de confusão do melhor modelo (ajustar nome conforme resultado)
best_model_name = 'xgboost'
plot_confusion_matrix(
    y_test,
    results[best_model_name]['y_pred'],
    save_path='../reports/figures/confusion_matrix.png'
)
```

```python
# Salvar o modelo final escolhido
save_model(trained_models[best_model_name], '../models/best_model.pkl')
```
